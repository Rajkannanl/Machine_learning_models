```python
print(y_pred)
```

```python
print("Mse:", mean_squared_error(y_test, y_pred))
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("RMSE:", rmse)
```

```python
print("R² Score:", r2_score(y_test, y_pred))
```
